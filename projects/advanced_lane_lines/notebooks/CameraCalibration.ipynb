```python
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import ipywidgets as widgets
import logging
logging.basicConfig(level=logging.INFO)

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from lane_detector import CameraCalibration

cc = CameraCalibration(9, 6, '../data/camera_cal', store_output_images=True)
```

```python
uncorrected = cv2.imread('../data/camera_cal/calibration3.jpg')
annotated = mpimg.imread('../data/camera_cal/output/calibration3.jpg')
corrected = cc.undistort_image(uncorrected)

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
f.tight_layout()
ax1.axis('off')
ax1.set_title('Original Image', fontsize=20)
ax1.imshow(uncorrected)

ax2.axis('off')
ax2.set_title('Annotated Image', fontsize=20)
ax2.imshow(annotated)

ax3.axis('off')
ax3.set_title('Undistorted Image', fontsize=20)
ax3.imshow(corrected)
```
